```python
import pandas as pd
import numpy as np
import nltk, re, string
import matplotlib.pyplot as plt

from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import svm
```

```python
pd.set_option('display.max_rows', None)
```

## Load

Pemberian label akan dilakukan secara manual dan lexicon-based.
Di bawah ini akan dilakukan load data label manual dari txt

```python
data = pd.read_csv('nmf_result.csv')
data.head()
```

```python
data['nmf_topic'].value_counts()
```

```python
plt.figure()
plt.hist(data['nmf_topic'], bins=60)
plt.show()
```

### SVM

Data training 80% dan data testing 20%

```python
train_X, test_X, train_Y, test_Y = model_selection.train_test_split(data['processed_text'], data['nmf_topic'], test_size = 0.2, random_state = 0)
```

Dibuat dataframe baru yang berisi data lexicon training dan testing

```python
df_train80 = pd.DataFrame()
df_train80['processed_text'] = train_X
df_train80['nmf_topic'] = train_Y

df_test20 = pd.DataFrame()
df_test20['processed_text'] = test_X
df_test20['nmf_topic'] = test_Y
```

```python
df_train80
```

```python
df_test20
```

## TF IDF

```python
# TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect_8020 = TfidfVectorizer(max_features = 5000)
tfidf_vect_8020.fit(data['processed_text'])
train_X_tfidf_8020 = tfidf_vect_8020.transform(df_train80['processed_text'])
test_X_tfidf_8020 = tfidf_vect_8020.transform(df_test20['processed_text'])
```

```python
tfidf_vect_8020
```

Hasil dari TF-IDF training

```python
print(train_X_tfidf_8020)
```

Hasil dari TF-IDF testing

```python
print(test_X_tfidf_8020)
```

```python
print(train_X_tfidf_8020.shape)
print(test_X_tfidf_8020.shape)
```

```python
print(tfidf_vect_8020.vocabulary_)
```

### SVM

```python
from sklearn.svm import SVC

model = SVC(kernel='linear', random_state=42)
model.fit(train_X_tfidf_8020,train_Y)
```

```python
from sklearn.metrics import accuracy_score

predictions_SVM_8020 = model.predict(test_X_tfidf_8020)
test_prediction_8020 = pd.DataFrame()
test_prediction_8020['text'] = test_X
test_prediction_8020['topic'] = predictions_SVM_8020
SVM_accuracy_8020 = accuracy_score(predictions_SVM_8020, test_Y)*100
SVM_accuracy_8020 = round(SVM_accuracy_8020,1)
```

```python
test_prediction_8020
```

```python
SVM_accuracy_8020
```

Akurasi menggunakan model SVM sebesar 88.1%

```python
from sklearn.metrics import classification_report

print ("\nHere is the classification report:") 
print (classification_report(test_Y, predictions_SVM_8020))
```

## Akurasi train

```python
model.score(train_X_tfidf_8020,train_Y)
```

```python
model.score(test_X_tfidf_8020,test_Y)
```
